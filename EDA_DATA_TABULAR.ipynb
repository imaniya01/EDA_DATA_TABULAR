# **6. Feature Engineering (optional)**

```python
kategori_list = [
    "soto",
    "nasi kuning",
    "nasi krawu",
    "rawon",
    "rujak cingur",
    "lontong kupang",
    "sate klopo",
    "lontong balap",
    "penyetan",
    "kikil",
    "nasi cumi",
    "tahu tek"
]

def get_kategori(nama):
    nama_lower = str(nama).lower()
    for kategori in kategori_list:
        if kategori in nama_lower:
            return kategori
    return "lainnya"

df['kategori_makanan'] = df['nama_makanan'].apply(get_kategori)

df.head(10)
```

# **5. EDA**

## 5.1 Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['rating_makanan'] = df['rating_makanan'].str.replace(',', '.').astype(float)
df['rating_restaurant'] = df['rating_restaurant'].str.replace(',', '.').astype(float)

numerical_cols = ['price', 'rating_makanan', 'rating_restaurant']

for col in numerical_cols:
    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Distribusi {col}')

    plt.show()
```

* Distribusi Harga memperlihatkan bahwa sebagian besar menu dijual dengan harga terjangkau di kisaran Rp20.000 hingga Rp30.000, sementara makanan dengan harga premium di atas Rp60.000 jumlahnya sangat jarang atau terbatas.
* Distribusi Rating Makanan memiliki pola yang serupa dengan rating restoran, di mana tingkat kepuasan konsumen terhadap makanan sangat tinggi dengan nilai yang paling sering muncul berada di angka 4.8.
* Distribusi Rating Restoran menunjukkan bahwa mayoritas restoran memiliki kualitas yang sangat baik menurut pelanggan, dengan nilai yang menumpuk di angka 4.6 hingga 4.8 dan sangat sedikit restoran yang memiliki rating rendah.

## 5.2 Multivariate Analysis

```python
sns.boxplot(x='kategori_makanan', y='price', data=df)
plt.xticks(rotation=45)
plt.title('Harga per Kategori Makanan')
plt.show()
```

* Distribusi harga tertinggi dimiliki oleh Rawon dan Kikil karena memiliki rata-rata yang tinggi dibanding kategori lainnya.
* Distribusi harga stabil dimiliki oleh sate klopo, lontong kupang, dan lontong balap.
* Kategori lainnya (minuman) dan Nasi Kuning memiliki outliers yang paling banyak (variasi harga).
* Lontong balap menjadi kategori yang paling murah keran garis mediannya paling bawah di bandingkan kategori lainnya.

## 5.2 Correlation Matriks

```python
numerical_cols = ['price', 'rating_makanan', 'rating_restaurant']
correlation = df[numerical_cols].corr()
print(correlation)

plt.figure(figsize=(6,5))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Variabel Numerik')
plt.show()
```
